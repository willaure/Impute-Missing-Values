```python
# Step 3. Finding and Removing Duplicates
# Task 1: Identify duplicate rows in the dataset.
```

```python
import pandas as pd

# Load the dataset
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

# Task 1: Identify duplicate rows
print("üîç DUPLICATE ROW ANALYSIS")
print("="*50)

# 1. Count total duplicate rows (entire row duplicates)
total_duplicates = df.duplicated().sum()
print(f"\nTotal complete duplicate rows: {total_duplicates}")

# 2. Display examples if duplicates exist
if total_duplicates > 0:
    print("\nSample duplicate rows (first 5):")
    duplicates = df[df.duplicated(keep=False)]  # keep=False marks all duplicates
    display(duplicates.sort_values(by=df.columns.tolist()).head())
else:
    print("\nNo complete duplicate rows found.")

# 3. Check for partial duplicates (based on selected columns)
key_columns = ['MainBranch', 'Employment', 'RemoteWork']
partial_duplicates = df.duplicated(subset=key_columns).sum()
print(f"\nPartial duplicates (based on {key_columns}): {partial_duplicates}")

# 4. Display partial duplicate examples
if partial_duplicates > 0:
    print("\nSample partial duplicates (first 5):")
    partial_dup_rows = df[df.duplicated(subset=key_columns, keep=False)]
    display(partial_dup_rows.sort_values(by=key_columns).head())
else:
    print("\nNo partial duplicates found based on selected columns.")

# 5. Show duplicate statistics
if total_duplicates > 0 or partial_duplicates > 0:
    print("\nüìä DUPLICATE STATISTICS:")
    dup_stats = pd.DataFrame({
        'Duplicate Type': ['Complete Rows', f'Partial ({", ".join(key_columns)})'],
        'Count': [total_duplicates, partial_duplicates],
        'Percentage': [
            total_duplicates/len(df)*100,
            partial_duplicates/len(df)*100
        ]
    })
    display(dup_stats.style.format({'Percentage': '{:.2f}%'}))
```

```python
##### Task 2: Remove the duplicate rows from the dataframe.
```

```python
import pandas as pd

# Load the dataset
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

# Task 2: Remove duplicate rows
print("üßπ DUPLICATE REMOVAL PROCESS")
print("="*50)

# Store original shape
original_shape = df.shape

# 1. Remove complete row duplicates (keeping first occurrence)
df_cleaned = df.drop_duplicates(keep='first')

# 2. Remove partial duplicates based on key columns
key_columns = ['MainBranch', 'Employment', 'RemoteWork']
df_cleaned = df_cleaned.drop_duplicates(subset=key_columns, keep='first')

# Calculate removed rows
complete_dups_removed = original_shape[0] - df.drop_duplicates(keep='first').shape[0]
partial_dups_removed = df.drop_duplicates(keep='first').shape[0] - df_cleaned.shape[0]
total_removed = original_shape[0] - df_cleaned.shape[0]

# Print results
print(f"\nOriginal dataset size: {original_shape}")
print(f"Cleaned dataset size: {df_cleaned.shape}")
print(f"\nRows removed:")
print(f"- Complete duplicates: {complete_dups_removed}")
print(f"- Partial duplicates: {partial_dups_removed}")
print(f"- TOTAL REMOVED: {total_removed} rows ({total_removed/original_shape[0]:.1%})")

# Verification
print("\n‚úÖ VERIFICATION:")
print(f"Remaining complete duplicates: {df_cleaned.duplicated().sum()}")
print(f"Remaining partial duplicates: {df_cleaned.duplicated(subset=key_columns).sum()}")

# Optional: Save cleaned data
# df_cleaned.to_csv('cleaned_survey_data.csv', index=False)
# print("\nCleaned data saved to 'cleaned_survey_data.csv'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

# Task 3: Find missing values for all columns
print("üîç COMPREHENSIVE MISSING VALUE ANALYSIS")
print("="*50)

# 1. Calculate missing values statistics
missing_data = pd.DataFrame({
    'Missing Values': df.isnull().sum(),
    'Percentage (%)': (df.isnull().mean() * 100).round(2),
    'Data Type': df.dtypes
}).sort_values('Missing Values', ascending=False)

# 2. Display columns with missing values
print("\nüìä MISSING VALUES SUMMARY:")
display(missing_data[missing_data['Missing Values'] > 0])

# 3. Visualize missing values
plt.figure(figsize=(14, 8))
sns.heatmap(df.isnull(),
            cbar=False,
            cmap=['#2ecc71', '#e74c3c'],  # Green=Present, Red=Missing
            yticklabels=False)
plt.title("Missing Values Heatmap", fontsize=16, pad=20)
plt.xticks(rotation=45, ha='right')
plt.show()

# 4. Detailed column analysis
print("\nüßê COLUMN-SPECIFIC ANALYSIS:")
for col in missing_data[missing_data['Missing Values'] > 0].index:
    print(f"\nColumn: {col} ({df[col].dtype})")
    print(f"Missing: {missing_data.loc[col, 'Missing Values']} ({missing_data.loc[col, 'Percentage (%)']}%)")

    if df[col].dtype == 'object':
        print("Top values:")
        print(df[col].value_counts(dropna=False).head(3).to_dict())
    else:
        print("Statistics:")
        print(df[col].describe().round(2).to_dict())

# 5. Overall summary
total_missing = df.isnull().sum().sum()
print(f"\nüìå TOTAL MISSING VALUES: {total_missing} ({total_missing/df.size:.1%} of all data)")
```

```python
# Task 6: Identify the most frequent (majority) value in the RemoteWork column.
```

```python
import pandas as pd

# Load the dataset
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

# Task 6: Find the most frequent value in RemoteWork
if 'RemoteWork' in df.columns:
    # Get the mode (most frequent value)
    mode_result = df['RemoteWork'].mode()

    # Handle cases where there might be multiple modes
    if len(mode_result) == 1:
        most_frequent = mode_result[0]
        print(f"The most frequent value in 'RemoteWork' is: '{most_frequent}'")

        # Show supporting statistics
        value_counts = df['RemoteWork'].value_counts()
        print("\nFull value distribution:")
        print(value_counts)

        # Calculate percentage
        percentage = (value_counts[most_frequent] / len(df)) * 100
        print(f"\n'{most_frequent}' appears in {percentage:.1f}% of responses")
    else:
        print("There are multiple modes in 'RemoteWork':")
        print(mode_result.tolist())
        print("\nThis means these values appear with the same highest frequency")
else:
    print("Error: 'RemoteWork' column not found in dataset")
    print("Available columns:", list(df.columns))
```

```python
# Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.¬∂
```

```python
import pandas as pd

# Load the dataset
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

# Task 7: Impute missing RemoteWork values with majority value
print("üîÑ REMOTEWORK IMPUTATION PROCESS")
print("="*50)

if 'RemoteWork' in df.columns:
    # 1. Check initial missing values
    missing_before = df['RemoteWork'].isna().sum()
    print(f"Missing values before imputation: {missing_before}")

    # 2. Find majority value (mode)
    majority_value = df['RemoteWork'].mode()[0]  # [0] takes first mode if multiple exist
    print(f"\nMajority value found: '{majority_value}'")

    # 3. Impute missing values
    df['RemoteWork'].fillna(majority_value, inplace=True)

    # 4. Verify results
    missing_after = df['RemoteWork'].isna().sum()
    print(f"\nMissing values after imputation: {missing_after}")

    # 5. Show new distribution
    print("\nUpdated value distribution:")
    print(df['RemoteWork'].value_counts())

    # 6. Calculate imputation impact
    imputed_count = missing_before - missing_after
    print(f"\nSuccessfully imputed {imputed_count} missing values")

    # Optional: Visual comparison
    import matplotlib.pyplot as plt

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    df['RemoteWork'].value_counts(dropna=False).plot(kind='bar', color='skyblue')
    plt.title('Before Imputation')
    plt.xticks(rotation=45)

    plt.subplot(1, 2, 2)
    df['RemoteWork'].value_counts().plot(kind='bar', color='lightgreen')
    plt.title('After Imputation')
    plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()

else:
    print("Error: 'RemoteWork' column not found")
    print("Available columns:", list(df.columns))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

# Task 8: Analyze compensation-related columns
print("üí∞ COMPENSATION ANALYSIS")
print("="*50)

# 1. Identify compensation-related columns
comp_cols = [col for col in df.columns if 'comp' in col.lower() or 'salary' in col.lower()]
print(f"Found {len(comp_cols)} compensation-related columns:")
print(comp_cols)

if not comp_cols:
    print("\nNo compensation-related columns found")
else:
    # 2. Analyze each compensation column
    for col in comp_cols:
        print(f"\nüîç Analyzing column: {col}")
        print("-"*40)

        # Basic stats
        missing = df[col].isna().sum()
        print(f"Missing values: {missing} ({missing/len(df):.1%})")
        print(f"Data type: {df[col].dtype}")

        # Distribution stats
        print("\nDescriptive statistics:")
        display(df[col].describe().to_frame().style.format("{:,.2f}"))

        # Visualize distribution
        plt.figure(figsize=(12, 5))

        # Histogram
        plt.subplot(1, 2, 1)
        sns.histplot(df[col].dropna(), bins=30, kde=True, color='royalblue')
        plt.title(f'Distribution of {col}')
        plt.xlabel('Amount')
        plt.ylabel('Frequency')

        # Boxplot
        plt.subplot(1, 2, 2)
        sns.boxplot(x=df[col], color='lightgreen')
        plt.title(f'Spread of {col}')
        plt.xlabel('Amount')

        plt.tight_layout()
        plt.show()

        # Top values
        if df[col].nunique() < 20:  # Only for categorical compensation fields
            print("\nTop values:")
            print(df[col].value_counts().head(10))
```
